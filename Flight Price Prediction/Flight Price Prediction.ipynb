```python
# Date of journey column is in object data type so need to be change it to datetime datatyoe.
```

```python
#converting into datetime format

#day
df_train['Journey_Day'] = pd.to_datetime(df_train.Date_of_Journey, format="%d/%m/%Y").dt.day

#month
df_train['Journey_month'] = pd.to_datetime(df_train["Date_of_Journey"], format="%d/%m/%Y").dt.month
```

```python
df_train.sample(10)
```

```python
#datatype of Date of journey column is comverted into new columns format
# and now droping Date of journey column. 
```

```python
df_train.drop(columns=['Date_of_Journey'],inplace=True)
```

```python
df_train.head()
```

```python
# converting Dep_time column into datetime format and again deleting the original column(Dep_time)
```

```python
#Hour
df_train['Dep_hour'] = pd.to_datetime(df_train['Dep_Time']).dt.hour

#Minute
df_train['Dep_min'] = pd.to_datetime(df_train['Dep_Time']).dt.minute
```

```python
#droping Dep_Time column
df_train.drop(columns=['Dep_Time'], inplace=True)
```

```python
df_train.head()
```

```python
#arrival time 
#Hour
df_train['Arrival_Hour'] = pd.to_datetime(df_train['Arrival_Time']).dt.hour

#Minute
df_train['Arrival_Min'] = pd.to_datetime(df_train['Arrival_Time']).dt.minute
```

```python
#droping arrival_time column
df_train.drop(columns=['Arrival_Time'],inplace=True)
```

```python
df_train.head()
```

```python
# duration column

#minutes
df_train['Duration_in_min'] = (pd.to_timedelta(df_train['Duration']).dt.seconds // 60).astype(int)
```

```python
#droping duration column
df_train.drop(columns=['Duration'], inplace=True)
```

```python
df_train.head()
```

```python
df_train.nunique()
```

```python
# converting dummies

Airline = pd.get_dummies(df_train[['Airline']],drop_first=True)
```

```python
Source = pd.get_dummies(df_train[['Source']],drop_first=True)
```

```python
Destination = pd.get_dummies(df_train[['Destination']],drop_first=True)
```

```python
Airline.head()
```

```python
Source.head()
```

```python
Destination.head()
```

```python
#merging all the dataframes

df = pd.concat([df_train, Airline, Source, Destination],axis=1)
```

```python
df.shape
```

```python
df
```

```python
df.drop(['Airline','Source','Destination'], axis=1, inplace=True)
```

```python
df
```

```python
#manipulating values of total stop column

df.replace({'non-stop': 0, '1 stop': 1, '2 stops':2, '3 stops':3, '4 stops':4}, inplace=True)
```

```python
df
```

```python
df.drop(['Route','Additional_Info'], axis=1, inplace=True)
```

```python
#checking correlation

plt.figure(figsize=(25,25))
sns.heatmap(df.corr(), annot=True, linecolor='blue', cbar=True)
plt.show()
```

```python
x = df.drop(['Price'], axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 7)
```

```python
#RANDOM FOREST MODEL
from sklearn.ensemble import RandomForestRegressor
reg_rf = RandomForestRegressor()
reg_rf.fit(x_train, y_train)
y_pred = reg_rf.predict(x_test)
```

```python
reg_rf.score(x_test, y_test)
```

```python
sns.distplot(y_test-y_pred)
plt.show()
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.show()
```

```python
#metrics
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt (metrics.mean_absolute_error(y_test, y_pred)))
```

```python
metrics.r2_score(y_test,y_pred)
```

```python
# Decision tree

from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

DecisionTreeModel = DecisionTreeClassifier(random_state=1)
DecisionTreeModel.fit(x_train, y_train)
y_pred = DecisionTreeModel.predict(x_test)
```

```python
DecisionTreeModel.score(x_test, y_test)
```

```python
#SVM

from sklearn import svm
from sklearn.metrics import accuracy_score

classifier = svm.SVC(kernel='linear')
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
```

```python
#score
classifier.score(x_test, y_test)
```
